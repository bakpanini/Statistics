# Central Limit Theorem

```python
from numpy import random
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

x = random.exponential(scale=1, size=1000)

sns.displot(x, stat="density", common_norm=False, kde=True)

plt.show()
```

```python
x.mean()
```

```python
# Form 1000 sets of 50 elements each and compute mean of each set. Store that mean in avg[]

avg = []
for _ in range(1000):
    sets = []
    set_size = random.randint(2, 100)
    for _ in range(set_size):
        idx = random.randint(0, len(x)-1)
        sets.append(x[idx])
    sets_mean = sum(sets)/len(sets)
    avg.append(sets_mean)
    
        
```

```python
sns.displot(avg, stat="density", common_norm=False, kde=True)
plt.title("Mean of sets gives a Gaussian distribution proving central limit theorem")
plt.show()
```

Notice we sampled from different sets from a exponential distribution and we computing mean of each of those sets and plotting gave us a **Guassian distribution**
