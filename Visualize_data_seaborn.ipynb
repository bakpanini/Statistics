```python
import seaborn as sns
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
titanic = sns.load_dataset("titanic")
```

```python
tips.head()
```

# Scatter Plot

```python
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="day", size="size",palette="YlGnBu")
plt.title("Scatter Plot")
```

# Histogram Plot

To see the **distribution of a random variable**. Here to see the distribution of a particuar column in dataset. <br>
kde here gives the curve.

```python
sns.histplot(tips['tip'], kde = True, bins = 20)
plt.title('Tip distribution - Histogram')
```

```python
sns.displot(tips['tip'], kde = True, bins = 20)
plt.title('Tip distribution - Histogram')
```

# Bar Graph

```python
sns.barplot(x="sex", y="tip", data=tips)
```

# Box Plot

```python
sns.boxplot(x="day", y="tip", data=tips, hue="sex", palette = "YlGnBu")
plt.title("Box Plot")
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", palette = "YlGnBu")
plt.title("Box Plot")
```

# Joint Plot

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind = "reg", palette = "YlGnBu")
```

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind = "kde", fill = True, cmap = "YlGnBu")
```
